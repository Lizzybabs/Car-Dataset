From the bar chart above, the largest engine type is size 6.0(Make=Cadillac, Type=Truck, Origin=USA	6.0)

```python
Origin_produce_max.plot.scatter(x = 'Type', y = 'EngineSize',s = 50, figsize=(11,6))
plt.subplots_adjust(wspace=5, hspace=2)
```

From the above scatter plot, sedan had the highest number of production

```python
Origin_produce_max.max()
```

```python
plt.figure(figsize=(6,6))
Origin_produce_max_counts = Origin_produce_max['Type'].value_counts()
plt.pie(Origin_produce_max_counts, labels = Origin_produce_max_counts.index, autopct='%1.1f%%')
plt.title('Amount of Cars Produced ')
```
